```python
from faker import Faker
import pandas as pd
import random
import numpy as np
```

```python
f = Faker()

customer_id = []
first_name =[]
last_name =[]
customer_email =[]
customer_phone =[]
customer_address =[]
state =[]
city =[]
pincode =[]


for _ in range(5000):
    customer_id.append(f.uuid4())
    first_name.append(f.first_name())
    last_name.append(f.last_name())
    customer_email.append(f.email())
    customer_phone.append(f.phone_number())
    customer_address.append(f.address())
    state.append(f.state())
    city.append(f.city())
    pincode.append(f.zipcode())

df_customer = pd.DataFrame(zip(customer_id,first_name,last_name,customer_email,customer_phone, customer_address, state, city, pincode),
columns=("customer_id","first_name", "last_name", "customer_email", "customer_phone", "customer_address", "state", "city", "pincode"))

df_customer
```

```python
customer_faker = "customer_dummydata.csv"
df_customer.to_csv(customer_faker, index=False)
```
